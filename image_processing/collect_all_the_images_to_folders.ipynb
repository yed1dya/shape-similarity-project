```python
from google.colab import drive
import os
import shutil

# Mount Google Drive
drive.mount('/content/drive')

# Path to the main folder
main_folder_path = '/content/drive/MyDrive/sample shapes/children'
shapes_folder_path = '/content/drive/MyDrive/sample shapes/shapes'

# Create folders for each shape number (1 to total_images)
def create_image_folders(shapes_folder_path, total_images=21):
    for i in range(1, total_images + 1):
        folder_path = os.path.join(shapes_folder_path, f'shape{i}')
        os.makedirs(folder_path, exist_ok=True)

# Collect and rename only MNIST-like images into respective folders
def collect_and_rename_mnist_images(main_folder_path, shapes_folder_path, total_images=21):
    for child_folder in os.listdir(main_folder_path):
        child_folder_path = os.path.join(main_folder_path, child_folder)

        # Ensure it's a directory
        if os.path.isdir(child_folder_path):
            # Path to the "Extracted images" folder
            extracted_images_folder = os.path.join(child_folder_path, "Extracted images")

            # Check if "Extracted images" folder exists
            if os.path.exists(extracted_images_folder):
                # Process each PNG file in "Extracted images"
                for file_name in os.listdir(extracted_images_folder):
                    if file_name.endswith(".png") and "mnist" in file_name:  # Only MNIST-like files
                        try:
                            # Extract the image number from the filename (e.g., '1mnist.png')
                            image_number = int(file_name.split('mnist')[0])  # Get the number before 'mnist'
                            if 1 <= image_number <= total_images:
                                # Target folder for this image
                                target_folder = os.path.join(shapes_folder_path, f'shape{image_number}')
                                target_file_name = f"{child_folder}.png"  # Rename to child folder name
                                target_file_path = os.path.join(target_folder, target_file_name)

                                # Source file path
                                source_file_path = os.path.join(extracted_images_folder, file_name)

                                # Copy the MNIST-like file to the target folder with the new name
                                shutil.copy(source_file_path, target_file_path)
                                print(f"Copied {file_name} from {child_folder_path} to {target_file_path}")
                        except ValueError:
                            print(f"Skipping {file_name} in {child_folder_path} (invalid filename format).")
            else:
                print(f"No 'Extracted images' folder found in {child_folder_path}.")
        else:
            print(f"{child_folder_path} is not a folder. Skipping.")

# Main function
def main():
    # Step 1: Create folders for each shape number
    create_image_folders(shapes_folder_path)

    # Step 2: Collect and organize MNIST-like images into the respective folders
    collect_and_rename_mnist_images(main_folder_path, shapes_folder_path)

# Run the main function
main()
```
